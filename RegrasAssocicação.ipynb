```python
dataset = [
    ["Leite", "Cebola", "Uva", "Feijão", "Ovos", "Iogurte"],
    ["Tomate", "Cebola", "Uva", "Feijão", "Ovos", "Iogurte"],
    ["Leite", "Maçã", "Feijão", "Ovos"],
    ["Leite", "Arroz", "Milho", "Feijão", "Tomate"],
    ["Milho", "Feijão", "Tomate", "Arroz", "Uva", "Ovos"]
]
```

```python
type(dataset)
```

```python
import pandas as pd 
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te
```

```python
te_dataset = te.fit(dataset).transform(dataset)
te_dataset
```

```python
df = pd.DataFrame(te_dataset, columns=te.columns_)
df
```

```python
te_dataset_2 = te.fit(dataset).transform(dataset, sparse=True)
te_dataset_2
```

```python
sparse_df = pd.DataFrame.sparse.from_spmatrix(te_dataset_2, columns=te.columns_)
sparse_df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
df
```

```python
apriori(df)
```

```python
apriori(df, use_colnames=True)
```

```python
apriori(df, min_support=0.8, use_colnames=True)
```

```python
itemset_frequentes = apriori(df, min_support=0.6, use_colnames=True)
itemset_frequentes['length'] = itemset_frequentes['itemsets'].apply(lambda x: len(x))
itemset_frequentes
```

```python
itemset_frequentes[(itemset_frequentes['length'] == 2) &
                   (itemset_frequentes['support'] >= 0.7)
                  ]
```

```python
itemset_frequentes[itemset_frequentes['itemsets'] == {'Uva', 'Feijão'}]
```

```python
itemset_frequentes[itemset_frequentes['itemsets'] == {'Tomate', 'Uva'}]
```
